```python
ports=pd.get_dummies(dataset.Embarked, prefix="Embarked",dtype="int")
```

```python
dataset=dataset.join(ports)
```

```python
titanic=dataset.drop(columns="Embarked")
```

```python
titanic
```

```python
titanic.Sex=titanic.Sex.map({'male':0,'female':1})
```

```python
titanic
```

```python
y=titanic.Survived.copy()
x=titanic.drop(columns="Survived")
```

```python
columns_to_drop=['Cabin','Ticket','Name','PassengerId']
```

```python
x=x.drop(columns=columns_to_drop)
```

```python
x
```

```python
x.isnull().values.any()
```

```python
x[pd.isnull(x).any(axis=1)] #searching for values which are zero (axis=1 means vertically)
```

```python
x.Age.fillna(x.Age.mean(),inplace=True)
```

```python
x #This completes the dataset which will now be trained using machine learning
```

```python
from sklearn.model_selection import train_test_split # importing the sklearn module's train_test_split
```

```python
x_train,x_valid,y_train,y_valid=train_test_split(x,y,test_size=0.2,random_state=8) #Training the model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_valid,y_valid)
```
